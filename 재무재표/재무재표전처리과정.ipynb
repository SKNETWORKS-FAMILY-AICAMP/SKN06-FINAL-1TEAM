```python
import pandas as pd
import os 

df = pd.read_csv("단일회사 전체 재무제표_OFS_2023.csv")
df = df.drop(columns=['status', 'message', "당기누적금액", "전기명(분/반기)", "전기금액(분/반기)", "전기누적금액","재무제표명","당기명","전기명","전전기명","보고서코드","고유번호"])
df.rename(columns={
    "당기금액": "2023",
    "전기금액": "2022",
    "전전기금액": "2021"
}, inplace=True)
categorys = ['BS', 'CIS', 'CF', 'SCE', 'IS']
output_folder = "Financial"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for category in categorys:
    df_category = df[df['재무제표구분'] == category]
    csv_filename = f"{output_folder}/{category}.csv"
    df_category.to_csv(csv_filename, index=False, encoding='utf-8-sig')
    print(f"{csv_filename} 파일이 생성")
```

# 재무상태표

```python
import pandas as pd
# BS : 재무상태표 IS : 손익계산서 CIS : 포괄손익계산서 CF : 현금흐름표 SCE : 자본변동표
BS = pd.read_csv("Financial/BS.csv")
# 당기금액         3593
# 전기명           960
# 전기금액         4523
# 전전기명         1849
# 전전기금액        5773
columns = ["당기금액", "전기명", "전기금액", "전전기명", "전전기금액"]
for col in columns:
    BS.fillna(0, inplace=True)
BS = BS.drop(columns=["계정 상세", "재무제표구분"])
BS["crop_code"] = BS["crop_code"].astype(str).str.zfill(8)  # 8자리 유지 (예시)

BS.to_csv("Financial/BS_cleaned.csv", index=False, encoding="utf-8-sig", quotechar='"')
print(BS[["crop_code"]].head())
BS
```
